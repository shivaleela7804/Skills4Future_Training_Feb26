```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# load the dataset
data = pd.read_csv('appliance_energy.csv')
data.head()
```

```python
# missing values?
data.isnull().sum()
```

```python
# Feature and target?
X = data[['Temperature (C)']]
y = data['Energy Consumption (kWh)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 42)
```

```python
# Instantiate model
model = LinearRegression()
```

```python
# train
model.fit(X_train, y_train)
```

```python
# predictions
y_pred = model.predict(X_test)
y_pred
```

```python
# metrics - mse & r2
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
r2 = r2_score(y_test, y_pred)
print(f'R Squared Score:{r2*100:.2f}%')
```

```python
plt.scatter(X_test,y_test,color='blue', label='Test Data') # Actual data
plt.scatter(X_test,y_pred,color='red', label='Regression Line') # Prediction data
plt.xlabel('Temperature (C)')
plt.ylabel('Energy Consumption (kWh)')
plt.legend()
plt.title('Energy Consumption Prediction using Simple Linear Regression')
plt.show()
```

```python
import joblib
joblib.dump(model,'SLRModel.pkl')
```

```python
# import necessary libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# load the dataset
gt = pd.read_csv('green_tech_data.csv')
gt.head()
```

```python
# Feature and target selection
gt.columns
```

```python
x = gt[['carbon_emissions', 'energy_output', 'renewability_index',
       'cost_efficiency']]
y = gt['sustainability']
```

```python
# Split the dataset
xtr, xte, ytr, yte = train_test_split(x, y, test_size = 0.2, random_state = 11)
```

```python
yte.shape
```

```python
xtr.shape
```

```python
#Instantiate the model
lr_model = LogisticRegression()
```

```python
# Training
lr_model.fit(xtr, ytr)
```

```python
# Predictions
lr_predict = lr_model.predict(xte)
lr_predict
```

```python
# Evaluation - metrics - accuracy , cr, cm
acc = accuracy_score(yte, lr_predict)
print(f'The accuracy score:{acc*100:.2f}%')
```

```python
cm = confusion_matrix(yte, lr_predict)
cm
```

```python
cr = classification_report(yte, lr_predict)
print(cr)
```
